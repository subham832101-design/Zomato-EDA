```python
# Group and sum votes
grouped_data = data.groupby('listed_in(type)')['votes'].sum()

# Convert to DataFrame
result = pd.DataFrame({'votes': grouped_data})

# Plot
plt.plot(result, c="green", marker="o")

# Labels
plt.xlabel("Type of Restaurant", c="blue", size=30)
plt.ylabel("Votes", c="blue", size=30)

plt.show()


```

```python

data=data[data['rate'].notna()].copy()
data['rate']=data['rate'].astype(str).str.replace('/5','',regex=False)
data['rate']=data['rate'].astype(float)
```

```python
data.head()
```

```python


plt.hist(data['rate'],bins=5)
plt.title("rating distribution")
plt.show()
```

# the majority resturant receive ratings from 3.5 to 4

```python
data.head()
```

```python
couple=data['approx_cost(for two people)']
sns.countplot(x=couple,color="red")
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x="online_order",y="rate",color="red",data=data)
```

```python
pivot_table=data.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("online order")
plt.ylabel("listed_in(type)")
plt.show()
```
